```python
import cymetric
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

from cymetric import graphs
from cymetric import timeseries
```

```python
ev = cymetric.Evaluator(cymetric.dbopen("source.h5"))
```

```python
graphs.flow_graph(ev)
```

```python
trans = timeseries.transactions(ev, receivers=["str"])
trans['Time'] = trans['Time']/12.


ax = trans.plot('Time','Mass',linestyle='', marker='+', markersize='20' ,figsize = (20,12))
# Major ticks every 20, minor ticks every 5
import numpy as np
major_y_ticks = np.arange(0, 20, 1)

ax.set_yticks(major_y_ticks)
major_x_ticks = np.arange(0, 30, 1)
ax.set_xticks(major_x_ticks)



# And a corresponding grid
ax.grid(which='both')

plt.show()


#u5_frac.plot(x=u5_frac.Time, y=u5_frac.Mass)
```
